# Notebook 02 — Processamento e EDA

**Propósito:** deflacionar pelo IPCA, tratar outliers/lacunas, executar análise exploratória e gerar **todas as tabelas e figuras da Seção 5.1 do TCC**.

**Saída:** arquivos `.tex` em `tables/generated/` e `.pdf` em `figures/generated/` do TCC.

```python
from forecasting.config import load_config
from forecasting.plotting import setup_matplotlib_thesis
from forecasting import io as fio, eda

setup_matplotlib_thesis()
cfg = load_config()

monthly = fio.load_monthly_series(cfg)
print(f'Carregadas {len(monthly)} linhas, colunas: {list(monthly.columns)[:8]} ...')
```

```python
# Deflacionamento por IPCA, base = dezembro de 2025.
monthly_real = eda.deflate_by_ipca(
    monthly,
    base_month=cfg.ipca_base_month,
    value_columns=['iptu', 'iss'],  # colunas em minusculo; 'iss' = ISSQN
)
monthly_real.head()
```

```python
# Geracao de TODOS os artefatos da Secao 5.1 do TCC.
artifacts = []
artifacts.append(eda.descriptive_stats_table(cfg, monthly_real))
artifacts.append(eda.stationarity_table(cfg, monthly_real))
artifacts.append(eda.time_series_panel(cfg, monthly_real))
artifacts.append(eda.seasonal_subseries_panel(cfg, monthly_real))
artifacts.append(eda.stl_decomposition_panel(cfg, monthly_real))
artifacts.append(eda.acf_pacf_panel(cfg, monthly_real))
for a in artifacts:
    print(f'  + {a}')
print(f'\nGeradas {len(artifacts)} artefatos. Recompile o TCC para visualizar.')
```
